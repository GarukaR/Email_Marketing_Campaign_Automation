```python
def extract_contact(test_url: str, counter: int = 1) -> dict:
    """
    Extract contact information from a webpage.
    
    Args:
        test_url (str): The URL of the webpage to extract contact information from.
        counter (int, optional): A counter for logging purposes. Defaults to 1.
    
    Returns:
        dict: A dictionary containing the extracted contact information with keys:
            'company', 'contact_name', 'email', 'telephone', and 'tag'.
    """
    driver.get(test_url)
    time.sleep(10)
    
    # Dictionary to store element locators and their corresponding keys
    elements_to_extract = {
        'contact_name': {'method': By.CLASS_NAME, 'locator': 'info'},
        'company_name': {'method': By.CLASS_NAME, 'locator': 'name'},
        'email': {'method': By.CSS_SELECTOR, 'locator': 'a.icon-label.round.t-company-contact.email'},
        'telephone': {'method': By.XPATH, 'locator': '/html/body/main/div[1]/section[1]/aside/div/div[2]/a[4]/span'},
        'tag': {'method': By.CLASS_NAME, 'locator': 'tag'}
    }
    
    # Extract each element safely
    extracted_data = {}
    for key, locator_info in elements_to_extract.items():
        try:
            extracted_data[key] = driver.find_element(locator_info['method'], locator_info['locator']).text
        except Exception:
            extracted_data[key] = ""
    
    # Rename company_name to company for consistency with return dict
    extracted_data['company'] = extracted_data.pop('company_name')
    
    print(f"{counter} : {extracted_data['company']} - {extracted_data['email']} extracted")
    
    return extracted_data
```
